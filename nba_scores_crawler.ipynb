```python
import requests 
from bs4 import BeautifulSoup 
import pandas as pd 

# df = pd.read_html(str(table))[0]
```

```python
months = [
'october',
'november',
'december',
'january',
'february',
'march',
'april',
'may',
'june',]
```

```python
df_all = None
first_flag = True
```

```python
for year in range(2019, 1999, -1):
    for month in months:
        print(year, month)
        link = "https://www.basketball-reference.com/leagues/NBA_{}_games-{}.html".format(year, month)
        res = requests.get(link) 
        soup = BeautifulSoup(res.content,'lxml')
        try:
            table = soup.find_all('table')[0]
        except Exception as e:
            continue
            
        df = pd.read_html(str(table))[0]
        
        if first_flag:
            first_flag=False
            df_all = df
        else:
            df_all = df_all.append(df, ignore_index=True)
```

```python
df_all.columns
```

```python
df_all = df_all[['Date', 'Home/Neutral' , 'Visitor/Neutral','PTS', 'PTS.1']]
```

```python
df_all['total_pts'] = df_all['PTS'] + df_all['PTS.1']
```

```python
df_all['total_pts_odd'] = df_all['total_pts'] % 2 
```

```python
_, df_all['month'], df_all['year'] = df_all['Date'].str.split(',').str
```

```python
df_all['month'] = df_all['month'].str.strip()
df_all['month'] = df_all['month'].str[:3]
```
