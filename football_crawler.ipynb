```python
import requests 
from bs4 import BeautifulSoup 
import pandas as pd 
import os
pd.set_option('display.max_rows', 2000)
```

```python
df_country_link = pd.DataFrame()

for country in os.listdir('football_data/'):
    if ('.' in country) or ('Other' in country):
        continue
    link = "http://www.football-data.co.uk/{}m.php".format(country.lower())
    print(link)
    res = requests.get(link) 
    soup = BeautifulSoup(res.content,'lxml')
    links = soup.find_all('a')
    links = [link for link in links
             if link["href"].endswith("csv")]
    c_links = [(country, link.attrs['href']) for link in links]
    df_country_link = df_country_link.append(c_links)
```

```python
df_country_link.columns = ['Country', 'CSV_link']
```

```python
df_country_link.reset_index(inplace=True, drop=True)
```

```python
for idx, row in df_country_link.iterrows():
    
    fname = '_'.join(row['CSV_link'].split('/')[1:])
    if os.path.exists('football_data/{}/{}'.format(row['Country'], fname)):
        continue
    print(idx)
    link = "http://www.football-data.co.uk/{}".format(row['CSV_link'])
    print(link)
    res = requests.get(link)
    print(fname)
    with open('football_data/{}/{}'.format(row['Country'], fname), 'wb') as f:
        f.write(res.content)
```
